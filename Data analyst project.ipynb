```python
### Handle the null values 
orders_df = pd.read_csv('orders.csv',na_values=['Not Available', 'unknown'])
```

```python
orders_df['Ship Mode'].unique()
```

```python
orders_df.isna().sum()
```

```python
# drop null values
orders_df.dropna(inplace=True)
```

```python
orders_df.shape
```

```python
### Rebname the column name
orders_df.columns
```

```python
orders_df.rename(columns={'Order Id':'order_id'})
```

```python
orders_df.columns = orders_df.columns.str.lower()
```

```python
orders_df.columns = orders_df.columns.str.replace(' ', '_')
orders_df
```

```python
# Adding new columns 
orders_df.columns
```

```python
orders_df['discount'] = orders_df['list_price'] * orders_df['discount_percent'] * .01
```

```python
orders_df.head()
```

```python
orders_df['sales_price'] = orders_df['list_price'] - orders_df['discount']
```

```python
orders_df.head()
```

```python
orders_df['profit'] = orders_df['sales_price'] - orders_df['cost_price']
```

```python
orders_df.head()
```

```python
# covert order date - object to datetime data type
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'], format='%Y-%m-%d')
```

```python
orders_df.dtypes
```

```python
# drop unwanted column 
orders_df.drop(columns=['cost_price', 'list_price', 'discount_percent'], inplace=True)
```
